```
df=pd.read_csv("/j2021-JEEnius-grades (3).csv")
df.head()
df = df.replace('-', 0)
```

```
df=df.drop(["ID number","State","Started on","Completed"],axis=1)
df=df.drop([len(df)-1])
```

```
maxlen=len(df)
percentile=[]
for t in range(maxlen):
   percentile.append(((maxlen-t-1)/maxlen)*100)
df["Percentile"]=percentile
df
```

```
chem=df.iloc[0:len(df),5:30]
math=df.iloc[0:len(df),55:80]
phy=df.iloc[0:len(df),30:55]
carr=chem.astype(float).sum(axis=1)
df["Chemistry"]=carr
marr=math.astype(float).sum(axis=1)
df["Math"]=marr
parr=phy.astype(float).sum(axis=1)
df["Physics"]=parr

```

```
maxlen=len(df)
percentile=[]
for t in range(maxlen):
   percentile.append(((maxlen-t-1)/maxlen)*100)

df=df.sort_values(by="Chemistry",ascending=False)
df["Chem Percentile"]=percentile
df=df.sort_values(by="Math",ascending=False)
df["Math Percentile"]=percentile
df=df.sort_values(by="Physics",ascending=False)
df["Physics Percentile"]=percentile
```

```
def weak_subjects(df,x):
  farr1=[]
  farr2=[]
  for i in range(1):
    clist=df.iloc[0,5:30].astype(float).tolist()  
    plist=df.iloc[0,30:55].astype(float).tolist()
    mlist=df.iloc[0,55:80].astype(float).tolist()
    cdict={}
    for x in chem:
      cdict[x]=0
    for x in range(25):
      cdict[chem[x]]+=clist[x]
    cweak=[]
    counted=[]
    maxc=[]
    for x in chem:
      if x not in counted:
        counted.append(x)
        maxc.append(chem.count(x)*4)
        if cdict[x]<=maxc[-1]//2:
          cweak.append(str(x)+":"+str(cdict[x])+"/"+str(maxc[-1]))
    mdict={}
    for x in math:
     mdict[x]=0
    for x in range(25):
      mdict[math[x]]+=mlist[x]  
    mweak=[]
    counted=[]
    maxm=[]
    for x in math:
      if x not in counted:
        counted.append(x)
        maxm.append(math.count(x)*4)
        if mdict[x]<=maxm[-1]//2:
          mweak.append(str(x)+":"+str(mdict[x])+"/"+str(maxm[-1]))
    pdict={}
    for x in phy:
      pdict[x]=0
    for x in range(25):
      pdict[phy[x]]+=plist[x]
    pweak=[]
    counted=[]
    maxp=[]
    for x in phy:
      if x not in counted:
        counted.append(x)
        maxp.append(phy.count(x)*4)
        if pdict[x]<=maxp[-1]//2:
          pweak.append(str(x)+":"+str(pdict[x])+"/"+str(maxp[-1]))
    farr1.append([cweak,mweak,pweak])
  return(farr1)
```

```
def accuracy(df,x):
   crr=df.iloc[x,5:95].astype(float).tolist()
   right=crr.count(4.00)+crr.count(4.0)
   wrong=crr.count(-1.00)+crr.count(-1.0)
   acc=100*(right/(right+wrong))
   return(acc)
```
